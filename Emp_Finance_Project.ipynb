## Estimating Volatility Dynamics

To capture the changing uncertainty in financial markets, we estimate volatility using a variety of GARCH-type models:

- **GARCH (1,1)**: Captures persistence and clustering in volatility.
- **EGARCH**: Models asymmetry in volatility, where negative returns might have stronger impact.
- **TGARCH**: Captures leverage effects where negative shocks increase volatility more than positive ones.

The models are estimated using daily log returns for the S&P 500 and Russell 2000 indices. Conditional volatilities are annualized and plotted to visualize how market uncertainty evolves over time.

Model performance is evaluated through:
- **Mean Squared Error (MSE)** for forecast accuracy.
- **Likelihood Ratio Tests** for comparing nested models.
- Visual inspection of residuals and volatility fit.


To determine the optimal volatility specification for each index, we conducted a grid search over GARCH family models by varying the order parameters p and q  from 1 to 4. Each model was evaluated using the Akaike Information Criterion (AIC), with the best fitting model being selected.


```python
def fit_best_volatility_model(resid, model_type: str, symmetry: int):
    p_range = range(1, 5)
    q_range = range(1, 5)
    results = []

    for p in p_range:
        for q in q_range:
            try:
                model = arch_model(resid, vol=model_type, p=p, o=symmetry, q=q, dist='normal', rescale=False)
                fitted_model = model.fit(disp="off")
                results.append({
                    'p': p, 'q': q,
                    'model': model,
                    'fit': fitted_model,
                    'aic': fitted_model.aic,
                    'bic': fitted_model.bic
                })
            except Exception as e:
                print(f"Model {model_type}({p},{q}) failed: {e}")

    if not results:
        raise ValueError("No valid models were fitted.")

    best_fit = min(results, key=lambda x: x['aic'])
    return best_fit['fit']
```

Annualized Conditional Volatility: GARCH, EGARCH, TGARCH

```python
def get_annualized_volatility(index: pd.DataFrame, model_type: str, symmetry: int):
    model = sm.ARIMA(index['Return'], order=(1, 0, 1)).fit()
    resid = model.resid

    fitted = fit_best_volatility_model(resid, model_type, symmetry)
    annual_vol = fitted.conditional_volatility * np.sqrt(252)

    mod = "TGARCH" if model_type == "GARCH" and symmetry == 1 else model_type
    return mod, index['Date'], annual_vol
```

Plotting Annualized Volatility

```python
def plot_combined_volatilities(index: pd.DataFrame, index_name: str, smooth=False):
    models = [
        ("GARCH", 0, 'blue', '-', 'GARCH'),
        ("EGARCH", 0, 'orange', '-', 'EGARCH'),
        ("GARCH", 1, 'green', '-', 'TGARCH')
    ]

    plt.figure(figsize=(14, 6))
    for model_type, symmetry, color, style, label in models:
        label, dates, vol = get_annualized_volatility(index, model_type, symmetry)
        if smooth:
            vol = vol.rolling(window=10).mean()
        plt.plot(dates, vol, label=label, color=color, linestyle=style, alpha=0.8)

    plt.title(f'Annualized Conditional Volatility - {index_name}')
    plt.xlabel('Date')
    plt.ylabel('Volatility')
    plt.legend()
    plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.5)
    plt.tight_layout()
    plt.show()
```

```python
plot_combined_volatilities(spx, "S&P 500", smooth=True)
plot_combined_volatilities(rut, "Russell 2000", smooth=True)
```

To assess volatility dynamics we fit GARCH(1,1), EGARCH, and TGARCH models to daily return data and compared their annualized conditional volatility estimates. As shown in the plots above, the models consistently captured key volatility spikes associated with major market disruptions, including the 1987 crash, the 2008 global financial crisis, and the onset of the pandemic in 2020, across both indices. The broad shape of the volatility series was similar across GARCH type models, confirming that all three are effective in modeling volatility clustering and persistence. However, EGARCH and TGARCH often displayed sharper peaks or more prolonged responses to negative shocks, reflecting their ability to capture asymmetry and leverage effects, which the standard GARCH models cannot.

Notably, the Russell 2000 exhibited consistently higher and more persistent volatility than the S&P500. This was particularly evident during crisis periods, where small cap volatility rose faster and remained elevated for a longer period of time. Additionally, the Russell 2000 showed more frequent mid level volatility surges in the absence of major crises, highlighting its increased sensitivity to idiosyncratic risks and economic uncertainty. These patterns are consistent with the riskier nature of small cap stocks and suggest that models accounting for asymmetry (EGARCH, TGARCH) may be especially valuable in small cap volatility modeling.


To quantitatively evaluate volatility model performance, we compared GARCH(1,1), EGARCH, and TGARCH forecasts against realized volatility computed as a 30 day rolling standard deviation of returns.

Performance Evaluation: Mean Squared Error

```python
indices = {'SPX': spx, 'RUT': rut}
models = [('GARCH', 0), ('EGARCH', 0), ('GARCH', 1)]
mse_results = []

for index_name, index_data in indices.items():
    actual_vol = index_data['Return'].rolling(window=30).std() * np.sqrt(252)
    for model_type, symmetry in models:
        try:
            _, _, predicted_vol = get_annualized_volatility(index_data, model_type, symmetry)

            df = pd.DataFrame({'actual': actual_vol, 'predicted': predicted_vol}).dropna()
            mse = mean_squared_error(df['actual'], df['predicted'])

            model_label = 'TGARCH' if model_type == 'GARCH' and symmetry == 1 else model_type
            mse_results.append({'Index': index_name, 'Model': model_label, 'MSE': mse})
        except Exception as e:
            continue

mse_df = pd.DataFrame(mse_results)
mse_df = mse_df.pivot(index='Model', columns='Index', values='MSE').round(6)
print(mse_df)
```

- GARCH(1,1) consistently yielded the lowest MSE for both indices, indicating superior forecast accuracy.
- EGARCH performed slightly worse but remained competitive
- TGARCH showed the highest prediction error, particularly for the S&P500.

These results suggest that the additional asymmetry terms in EGARCH and TGARCH do not necessarily translate into improved short term forecasting accuracy in this context. However, their ability to capture non linear responses may still offer modeling value from a structural or interpretive standpoint.



Complementing the MSE evaluation, we performed Likelihood Ratio (LR) tests to compare the relative fit of the models.


Performance Evaluation: Likelihood Ratio Test (LR Test)

```python
resid = sm.ARIMA(spx['Return'], order=(1, 0, 1)).fit().resid

def lr_test(model1: str, model2: str, symmetry1=0, symmetry2=0):
    model1_fit = fit_best_volatility_model(resid, model1, symmetry1)
    model2_fit = fit_best_volatility_model(resid, model2, symmetry2)
    model1 = "TGARCH" if model1 == "GARCH" and symmetry1 == 1 else model1
    model2 = "TGARCH" if model2 == "GARCH" and symmetry2 == 1 else model2

    if model2_fit.num_params - model1_fit.num_params == 0:
        print(f"LR-test between {model1} and {model2} not applicable because df = 0")
        return
    elif model2_fit.num_params - model1_fit.num_params > 0:
        model1, model2 = model2, model1
        model1_fit, model2_fit = model2_fit, model1_fit

    lr_stat = 2 * (model1_fit.loglikelihood - model2_fit.loglikelihood)
    df = model1_fit.num_params - model2_fit.num_params
    p_value = 1 - chi2.cdf(lr_stat, df)
    if p_value < 0.05:
        print(f"H0 rejected: {model1} fits significantly better than {model2} (LR Stat: {lr_stat:.4f}, df: {df}, p-value: {p_value:.4f})")
    else:
        print(f"H0 not rejected: {model1} not significantly better than {model2} (LR Stat: {lr_stat:.4f}, df: {df}, p-value: {p_value:.4f})")


lr_test("GARCH", "EGARCH")
lr_test("GARCH", "GARCH", 0, 1)
lr_test("EGARCH", "GARCH", 0, 1)
```

The results indicate that the standard GARCH(1,1) model provides a significantly better fit than the EGARCH specification (LR Stat = 10.97 and p = 0.0041). While a formal comparison between GARCH and TGARCH was not applicable due to identical degrees of freedom, the TGARCH model was found to significantly outperform EGARCH (LR Stat = 281.90 and p < 0.0001). These results suggest that while EGARCH is theoretically well suited to capturing asymmetries, it underperforms both GARCH and TGARCH on empirical grounds for our case.

### Time-Varying Correlation

This section examines how the correlation between large- and small-cap stock returns changes over time, especially during market stress periods. We use three methods:

- **Rolling Window Correlation**: Uses a fixed-length window to compute Pearson correlation dynamically.
- **EWMA (Exponentially Weighted Moving Average)**: Places higher weight on recent observations to detect short-term co-movement.
- **DCC-GARCH (Dynamic Conditional Correlation GARCH)**: A multivariate GARCH model that captures time-varying correlations with volatility spillovers between assets.

We visualize each method to see how correlation behaves, especially during known financial crises (e.g., 2008, COVID-19).

The DCC-GARCH method is particularly useful for capturing dynamic relationships in volatility-adjusted environments.

Rolling Window Correlation

```python
def rolling_correlation(df1, df2):
    windows = [30, 90, 180]
    colors = ['#1f77b4', '#ff7f0e', '#2ca02c']

    plt.figure(figsize=(14, 6))

    for w, c in zip(windows, colors):
        rolling = df1['Return'].rolling(window=w).corr(df2['Return'])
        plt.plot(df1['Date'], rolling, label=f'{w}-Day Window', color=c, linestyle="-", linewidth=2)

    plt.title('Rolling Window Correlation (SPX vs RUT)', fontsize=14)
    plt.xlabel('Date')
    plt.ylabel('Correlation')
    plt.ylim(0, 1.05)
    plt.grid(True, linestyle='--', alpha=0.4)
    plt.legend(title='Window Size')
    plt.tight_layout()
    plt.show()
```

```python
rolling_correlation(spx, rut)
```

EWMA (Exponentially Weighted Moving Average)

```python
def ewma_cov_corr(df1, df2, lambda_=0.94):
    r1 = df1['Return']
    r2 = df2['Return']
    cov = (r1 * r2).ewm(alpha=1 - lambda_).mean()
    var1 = (r1 ** 2).ewm(alpha=1 - lambda_).mean()
    var2 = (r2 ** 2).ewm(alpha=1 - lambda_).mean()
    corr = cov / np.sqrt(var1 * var2)
    return corr

spx['EWMACorr'] = ewma_cov_corr(spx, rut)

plt.figure(figsize=(12, 4))
plt.plot(spx['Date'], spx['EWMACorr'], label='EWMA Correlation (=0.94)', color='orange')
plt.title('EWMA Time-Varying Correlation (SPX vs RUT)')
plt.xlabel('Date')
plt.ylabel('Correlation')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

Both plots reveal consistently strong positive correlation, with notable spikes during periods of financial turmoil. These increases are consistent with the well established phenomenon of rising cross asset correlation during systemic stress. While both methods successfully capture these dynamics, EWMA responds more rapidly to recent changes due to its exponential weighting scheme, making it more sensitive to sudden market shifts. In contrast, the rolling window approach provides a smoother trend, particularly with longer windows (180 days) while shorter windows (30 days) capture sharper variations in correlation but exhibit greater volatility. This method may also lag in detecting abrupt structural breaks.

## Lead-Lag Relationships

We investigate the possibility of volatility spillover between the S&P 500 and Russell 2000 using VAR (Vector Autoregression) models and Granger causality tests.

- **VAR Model**: Models the interdependence of squared returns (proxy for volatility) across time lags. Optimal lag length is chosen based on AIC/BIC.
- **Granger Causality Test**: Evaluates if past values of one series help forecast the other, indicating possible lead-lag dynamics.
- **Impulse Response Analysis**: Examines the effect of a one-time volatility shock in one index on the other over time.

These tools help us identify whether large-cap stocks tend to lead or follow small-cap stocks in volatility transmission, offering insights into market structure and risk.

Vector Autoregression (VAR)

```python
def garch_model(index: pd.DataFrame):
    model = sm.ARIMA(index['Return'], order=(1, 0, 1)).fit()
    resid = model.resid

    p_range = range(1, 5)
    q_range = range(1, 5)
    results = []

    for p in p_range:
        for q in q_range:
            try:
                model = arch_model(resid, vol="GARCH", p=p, q=q, dist='normal', rescale=False)
                fitted_model = model.fit(disp='off')
                results.append({'p': p, 'q': q, 'AIC': fitted_model.aic, 'BIC': fitted_model.bic})
            except Exception as e:
                continue

    results_df = pd.DataFrame(results)
    results_df.sort_values(by='AIC', inplace=True)
    best_model_params = results_df.iloc[0]

    best_model = arch_model(resid, vol="GARCH", p=int(best_model_params['p']), q=int(best_model_params['q']), dist='normal', rescale=False)
    return best_model.fit(disp='off')

spx_vol = garch_model(spx).conditional_volatility
rut_vol = garch_model(rut).conditional_volatility

vol_data = pd.DataFrame({'SPX_vol': np.log(spx_vol * np.sqrt(252)), 'RUT_vol': np.log(rut_vol * np.sqrt(252))}).dropna()

model = VAR(vol_data)
results = model.fit(maxlags=20, ic='aic')

print(results.summary())
```

Performance Evaluation: R Squared (VAR Model)

```python
def var_r_squared(var_results):
    r_squared = {}
    k_ar = var_results.k_ar
    endog = var_results.model.endog
    for i, endog_name in enumerate(var_results.model.endog_names):
        actual = endog[k_ar:, i]
        fitted = var_results.fittedvalues.iloc[:, i].values
        ss_res = np.sum((actual - fitted) ** 2)
        ss_tot = np.sum((actual - np.mean(actual)) ** 2)
        r_squared[endog_name] = 1 - (ss_res/ss_tot)
    return r_squared

r2_dict = var_r_squared(results)
print("R-squared values per equation:")
for var, r2 in r2_dict.items():
    print(f"{var}: {r2:.4f}")
```

The model fit was strong, with R values of 0.9738 for RUT_vol and 0.9658 for SPX_vol, this confirms the presence of persistent volatility patterns in both indices.

Granger Causality Tests

```python
# Test if SPX_vol Granger-causes RUT_vol
granger_test_spx = results.test_causality('RUT_vol', ['SPX_vol'], kind='f')
print(granger_test_spx.summary())

# Test if RUT_vol Granger-causes SPX_vol
granger_test_rut = results.test_causality('SPX_vol', ['RUT_vol'], kind='f')
print(granger_test_rut.summary())
```

Granger causality tests indicate a directional relationship in volatility dynamics: shocks to the Russell 2000 significantly Granger cause volatility in the S&P 500 (p < 0.001), whereas the reverse effect, from S&P 500 to Russell 2000, is only marginally significant (p = 0.052). This suggests that volatility in the small cap segment may act as a leading indicator for broader market volatility.


As seen in the first and last plot, a shock to SPX_vol or RUT_vol resulted in an immediate and gradually decaying effect on its own volatility, indicating strong persistence. Additionally, both indices exhibited self reinforcing volatility, with RUT_vol showing more persistent clustering than SPX_vol.
